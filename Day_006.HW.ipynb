```python
def ecdf(data):
    xs = np.sort(data)
    ys = np.arange(1, len(xs)+1)/float(len(xs))
    return xs, ys
```

```python
print(f_app_train['REGION_POPULATION_RELATIVE'].describe())

xs, ys = ecdf(f_app_train['REGION_POPULATION_RELATIVE'])
plt.figure(figsize=(14,4))
plt.plot(xs, ys,label="REGION_POPULATION_RELATIVE")
plt.title('ECDF of REGION_POPULATION_RELATIVE')
plt.xlabel('Value')
plt.ylabel('ECDF')
plt.xlim([xs.min(), xs.max()]) 
plt.ylim([-0.05,1.05]) 
plt.legend()
plt.show()

f_app_train['REGION_POPULATION_RELATIVE'].hist()
plt.show()

f_app_train['REGION_POPULATION_RELATIVE'].value_counts()
```
