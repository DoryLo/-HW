```python
from sklearn.model_selection import train_test_split, KFold
import numpy as np
```

```python
# 生成從 0 到 50 的 array，並 reshape 成 (10, 5) 的 matrix
X = np.arange(50).reshape(10, 5)
 # 生成一個全零 arrary
y = np.zeros(10)
# 將一半的值改為 1
y[:5] = 1
print("Shape of x :", X.shape)
print("Shape of y :", y.shape)
```

```python
print('X : shape  ' + str(X.shape))
print(X)
print("")
print('y : shape ' + str(y.shape))
print('y : shape : ' + str(y.shape))
print(y)
```

```python
X_train, X_test, y_train, y_trst = train_test_split(X, y, test_size = 2, random_state = 0)
```

```python
kf = KFold(n_splits = 5)
i = 0
for train_index, test_index in kf.split(X) :
    i +=1
    X_train, X_test = X[train_index], X[test_index]
    y_train, y_test = y[train_index], y[test_index]
    print("FOLD {} : ".format(i))
    print("X_test :", X_test)
    print("Y_test :", y_test)
    print("-"*30)
```
